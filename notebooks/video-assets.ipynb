# Generating Video Assets

This notebook is purely for generating assets for the video.

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams.update({'font.size': 12})
import numpy as np

from beat2d import settings
```

# Waveforms

```python
sample, sr = librosa.core.load("../data/test/002.wav", settings.SAMPLE_RATE)

sample, trim_pts = librosa.effects.trim(sample, top_db=30)
sample = sample[:librosa.time_to_samples(20, settings.SAMPLE_RATE)]

with plt.style.context("seaborn-dark-palette"):
    plt.figure(figsize=(13.8, 4.0), dpi=100, facecolor=(1, 1, 1))
    plt.title("Beatboxing Waveform")
    plt.xlabel("time (s)")
    plt.ylabel("amplitude")
    plt.plot(librosa.samples_to_time(range(len(sample))), sample)
```

# Onsets

```python
sample, sr = librosa.core.load("../data/test/002.wav", settings.SAMPLE_RATE)

sample, trim_pts = librosa.effects.trim(sample, top_db=30)
sample = sample[:librosa.time_to_samples(20, settings.SAMPLE_RATE)]

o_env = librosa.onset.onset_strength(sample, sr=sr)
times = librosa.times_like(o_env, sr=sr)
onset_frames = librosa.onset.onset_detect(onset_envelope=o_env, sr=sr, backtrack=True, units="frames")

with plt.style.context("seaborn-dark-palette"):
    plt.figure(figsize=(13.8, 4.0), dpi=100)
    plt.title("Onset Detection")
    plt.plot(times, o_env, label='Onset strength')
    plt.vlines(times[onset_frames], 0, o_env.max(), color='r', alpha=0.9,
               linestyle='--', label='Onsets')
    plt.xlabel("time (s)")
    plt.ylabel("amplitude (s)")
    plt.legend()
```
